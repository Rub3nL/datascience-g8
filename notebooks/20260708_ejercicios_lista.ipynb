```python
# Ejercio 01

numeros = [4, 7, 2, 4, 9, 7, 1, 2, 10, 4]
numeros_distict = []

for numero in numeros:
    if numero not in numeros_distict:
        numeros_distict.append(numero)

print(numeros_distict)
```

```python
# Ejercio 02

numeros = [4, 7, 2, 4, 9, 7, 1, 2, 10, 4]

numero_pares = [
    numero
    for numero in numeros
    if numero % 2 == 0
]

numero_impares = [
    numero
    for numero in numeros
    if numero % 2 == 1
]

print(f"numero_pares:  {numero_pares}")
print(f"numero_impares: {numero_impares}")
```

```python
# Ejercico 03

numeros = [45, 12, 89, 34, 89, 23, 67, 10]

mayor = 0
segundo_mayor = 0

for n in numeros:
    if n > mayor:
        segundo_mayor = mayor
        mayor = n
    elif n > segundo_mayor and n < mayor:
        segundo_mayor = n

print(f"El segundo número mayor es: {segundo_mayor}")
```

```python
# Ejercicio 04

palabras = ['python', 'sol', 'programacion', 'lista', 'computadora', 'if', 'variable']
contador = 0

for palabra in palabras:
    if len(palabra) > 5:
        contador += 1

print(f"Número de palabras con más de 5 letras: {contador}")
```

```python
# Ejercicio 05

notas = [14, 18, 11, 16, 9, 13, 20]

promedio = round(sum(notas) / len(notas),2)

nota_minima = 11

print(f"Promedio: {promedio}")

if promedio >= nota_minima:
    print("Estado: Aprobado")
else:
    print("Estado: Desaprobado")
```

```python
# Ejercicio 09

lista1 = [4, 8, 12, 16, 20]
lista2 = [8, 10, 12, 18, 20]

unicos = []

for x in lista1:
    if x not in lista2:
        unicos.append(x)

for x in lista2:
    if x not in lista1:
        unicos.append(x)

print(f"Elementos únicos entre ambas listas: {unicos}")
```

```python
# Ejercicio 10

colores = ['rojo', 'azul', 'verde', 'amarillo', 'morado']
colores_invertidos = []

i = len(colores) - 1

while i >= 0:
    colores_invertidos.append(colores[i])
    i -= 1

print(f"Lista invertida: {colores_invertidos}")
```

```python
# Ejercicio 11

palabras = ['casa', 'elefante', 'sol', 'computadora', 'mesa', 'programacion']

print(f"Lista original: {palabras}")

n = len(palabras)
for i in range(n):
    for j in range(0, n - i - 1):
        # Comparamos la longitud
        if len(palabras[j]) > len(palabras[j + 1]):
            # Si la actual es más larga, las intercambiamos
            palabras[j], palabras[j + 1] = palabras[j + 1], palabras[j]

print(f"Lista ordenada: {palabras}")
```

```python
# Ejercicio 12

textos = ['hola', '', 'python', '', 'listas', ' ', 'programar', '']
limpios = []

for t in textos:
    if t.strip() != '':
        limpios.append(t)

print(f"Lista original: {textos}")
print(f"Lista sin vacíos: {limpios}")
```

```python
# Ejercicio 13

nombres = ['Ana', 'Luis', 'Ana', 'Carlos', 'Luis', 'Ana', 'María', 'Carlos']
frecuencia = {}

for nombre in nombres:
    if nombre in frecuencia:
        frecuencia[nombre] += 1
    else:
        frecuencia[nombre] = 1

for nombre, cantidad in frecuencia.items():
    print(f"{nombre}: {cantidad}")
```

```python
# Ejercicio 15
frutas = ['manzana', 'pera', 'uva', 'sandía', 'plátano', 'mango']

fruta_buscada = input("Introduce el nombre de una fruta: ").lower()

if fruta_buscada in frutas:
    posicion = frutas.index(fruta_buscada)
    print(f"La fruta '{fruta_buscada}' está en la posición: {posicion + 1}")
else:
    print(f"Lo siento, la fruta '{fruta_buscada}' no se encuentra en la lista.")
```

```python
# Ejercicio 16
numeros = [10, 20, 30, 40, 50, 60, 70, 80]

punto_medio = len(numeros) // 2

parte1 = []
parte2 = []

for idx, numero in enumerate(numeros):
    if idx < punto_medio:
        parte1.append(numero)
    else:
        parte2.append(numero)

print(f"Parte 1: {parte1}")
print(f"Parte 2: {parte2}")
```

```python
# Ejercicio 17
palabras = ['java', 'python', 'java', 'c++', 'java', 'javascript']

print(f"Lista original: {palabras}")

for i in range(len(palabras)):
    if palabras[i] == 'java':
        palabras[i] = 'python'

print(f"Lista actualizada: {palabras}")
```
