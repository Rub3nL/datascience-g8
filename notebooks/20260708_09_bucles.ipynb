```python
nombre = "Ruben"
indice = 0

while indice < len(nombre):
    print(nombre[indice])
    indice += 1  # Incrementamos el índice para pasar a la siguiente letra
```

```python
nombre = "Ruben"
for letra in nombre:
    if letra == "e":
        break
    print(letra)
```

```python

numero = int(input("Ingresa el número: "))

for i in range(numero):
    print(f"Número FOR: {numero}")


contador = 1
while contador <= 10:
    print(f"Número WHILE: {numero}")
    contador += 1
```

```python

n = int(input("Ingresa un número: "))

resultado = 1
i = 1

while i <= n:
    resultado = resultado * i
    i = i + 1

print(f"El factorial de {n} es {resultado}")
```

```python
n = int(input("Ingresa un número: "))

resultado = 1

for i in range(1, n + 1):
    resultado = resultado * i

print(f"El factorial de {n} es {resultado}")
```
