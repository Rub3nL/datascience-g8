```python
# Definición de los conjuntos
estudiantes_python = {"Ana", "Luis", "María", "Carlos", "Sofía", "Diego"}
estudiantes_javascript = {"María", "Carlos", "Pedro", "Lucía", "Diego", "Valeria"}

# 1. Estudiantes inscritos en ambos cursos (Intersección)
ambos = estudiantes_python & estudiantes_javascript

# 2. Estudiantes que solo están en Python (Diferencia)
solo_python = estudiantes_python - estudiantes_javascript

# 3. Estudiantes que solo están en JavaScript (Diferencia)
solo_javascript = estudiantes_javascript - estudiantes_python

# 4. Todos los estudiantes sin repetir (Unión)
todos = estudiantes_python | estudiantes_javascript

# Resultados
print(f"Estudiantes en ambos cursos: {ambos}")
print(f"Estudiantes solo en Python: {solo_python}")
print(f"Estudiantes solo en JavaScript: {solo_javascript}")
print(f"Todos los estudiantes: {todos}")
```
