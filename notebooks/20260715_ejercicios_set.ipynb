```python
# Ejercicio 01

estudiantes_python = {"Ana", "Luis", "María", "Carlos", "Sofía", "Diego"}
estudiantes_javascript = {"María", "Carlos", "Pedro", "Lucía", "Diego", "Valeria"}

# 1. Estudiantes inscritos en ambos cursos
ambos = estudiantes_python.intersection(estudiantes_javascript)
print(f"Inscritos en ambos cursos: {ambos}")

# 2. Estudiantes que solo están en Python
solo_python = estudiantes_python.difference(estudiantes_javascript)
print(f"Solo en Python: {solo_python}")

# 3. Estudiantes que solo están en JavaScript
solo_javascript = estudiantes_javascript.difference(estudiantes_python)
print(f"Solo en JavaScript: {solo_javascript}")

```

```python
# Ejercicio 02

almacen_arequipa = {"laptop", "teclado", "mouse", "monitor", "impresora", "audifonos"}
almacen_lima = {"laptop", "mouse", "monitor", "tablet", "webcam", "parlantes"}

# 1. Productos disponibles en ambos
comunes = almacen_arequipa.intersection(almacen_lima)
print(f"Productos en ambos almacenes: {comunes}")

# 2. Productos exclusivos de Arequipa
exclusivos_arequipa = almacen_arequipa.difference(almacen_lima)
print(f"Exclusivos de Arequipa: {exclusivos_arequipa}")

# 3. Productos exclusivos de Lima
exclusivos_lima = almacen_lima.difference(almacen_arequipa)
print(f"Exclusivos de Lima: {exclusivos_lima}")

# 4. Todos los productos disponibles sin repetir
inventario_total = almacen_arequipa.union(almacen_lima)
print(f"Total de productos únicos: {inventario_total}")

# 5. Determinar si los inventarios son iguales
son_iguales = almacen_arequipa == almacen_lima
print(f"¿Los inventarios son idénticos?: {son_iguales}")
```

```python
# Ejercicio 03

permisos_usuario = {"leer", "editar", "comentar", "descargar"}
permisos_requeridos = {"leer", "editar", "eliminar", "publicar"}

# 1. Si el usuario posee todos los permisos
tiene_todos = permisos_usuario.issuperset(permisos_requeridos)
print(f"¿Posee todos los permisos requeridos?: {tiene_todos}")

# 2. Permisos que le faltan
faltantes = permisos_requeridos.difference(permisos_usuario)
print(f"Permisos que le faltan: {faltantes}")

# 3. Permisos adicionales
adicionales = permisos_usuario.difference(permisos_requeridos)
print(f"Permisos adicionales que tiene: {adicionales}")

# 4. Cantidad total de permisos diferentes
total_permisos_unicos = len(permisos_usuario.union(permisos_requeridos))
print(f"Cantidad total de permisos diferentes: {total_permisos_unicos}")

```

```python
# Ejercicio 04

numeros_a = {2, 4, 6, 8, 10, 12, 14}
numeros_b = {4, 6, 9, 10, 13, 14, 16}
numeros_c = {1, 4, 6, 10, 11, 14, 18}

# 1. Números presentes en los tres conjuntos (Intersección)
en_los_tres = numeros_a.intersection(numeros_b).intersection(numeros_c)
print(f"En los tres conjuntos: {en_los_tres}")

# 2. Números en A y B, pero no en C (Intersección y Diferencia)
a_y_b_no_c = numeros_a.intersection(numeros_b).difference(numeros_c)
print(f"En A y B, pero no en C: {a_y_b_no_c}")

# 3. Números que solo pertenecen a A (Diferencia sucesiva)
solo_a = numeros_a.difference(numeros_b).difference(numeros_c)
print(f"Solo en A: {solo_a}")

# 4. Todos los números diferentes (Unión)
todos = numeros_a.union(numeros_b).union(numeros_c)
print(f"Todos los números presentes: {sorted(list(todos))}")
```

```python
# Ejercicio 05

habilidades_candidato_1 = {"Python", "Git", "SQL", "Docker", "React"}
habilidades_candidato_2 = {"Python", "JavaScript", "React", "Node.js", "Git"}
habilidades_requeridas = {"Python", "Git", "SQL", "Docker"}

# 1. Habilidades que comparten ambos candidatos
compartidas = habilidades_candidato_1.intersection(habilidades_candidato_2)
print(f"Habilidades compartidas: {compartidas}")

# 2. Habilidades exclusivas de cada uno
exclusivas_1 = habilidades_candidato_1.difference(habilidades_candidato_2)
exclusivas_2 = habilidades_candidato_2.difference(habilidades_candidato_1)
print(f"Exclusivas Candidato 1: {exclusivas_1}")
print(f"Exclusivas Candidato 2: {exclusivas_2}")

# 3. Habilidades requeridas que faltan
faltantes_1 = habilidades_requeridas.difference(habilidades_candidato_1)
faltantes_2 = habilidades_requeridas.difference(habilidades_candidato_2)
print(f"Faltantes Candidato 1: {faltantes_1}")
print(f"Faltantes Candidato 2: {faltantes_2}")

# 4. Determinar quién cumple más requisitos
cumplen_1 = habilidades_candidato_1.intersection(habilidades_requeridas)
cumplen_2 = habilidades_candidato_2.intersection(habilidades_requeridas)
print(f"\nCandidato 1 cumple {len(cumplen_1)} requisitos: {cumplen_1}")
print(f"Candidato 2 cumple {len(cumplen_2)} requisitos: {cumplen_2}")

# 5. Todas las habilidades disponibles
total = habilidades_candidato_1.union(habilidades_candidato_2)

if len(cumplen_1) > len(cumplen_2):
    print("\nEl Candidato 1 cumple con más requisitos.")
elif len(cumplen_2) > len(cumplen_1):
    print("\nEl Candidato 2 cumple con más requisitos.")
else:
    print("\nAmbos cumplen con la misma cantidad de requisitos.")

print(f"Total de habilidades únicas disponibles: {total}")
```
