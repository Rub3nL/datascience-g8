```python
# Ejercicio 01

inventario_norte = {"laptop": 4, "mouse": 10, "monitor": 3}
inventario_sur = {"laptop": 2, "teclado": 8, "monitor": 5}
inventario_centro = {"mouse": 6, "teclado": 4, "webcam": 7}

inventario_total = {}

for llave,valor in inventario_norte.items():
    if llave not in inventario_total:
        inventario_total[llave] = 0
    inventario_total[llave] += valor

for llave,valor in inventario_sur.items():
    if llave not in inventario_total:
        inventario_total[llave] = 0
    inventario_total[llave] += valor

for llave,valor in inventario_centro.items():
    if llave not in inventario_total:
        inventario_total[llave] = 0
    inventario_total[llave] += valor

print("Inventario Total original: ",inventario_total)

print("\t")
inventario_total = dict(sorted(inventario_total.items()))
print("Inventario Total ordenado: ",inventario_total)
```

```python
calificaciones = {
    "Ana": [17, 18, 16, 19],
    "Luis": [15, 16, 17, 16],
    "Carla": [18, 17, 19, 18],
    "Pedro": [16, 16, 16, 16],
    "Lucía": [15, 17, 16, 16]
}

# 1. Crear el diccionario con los datos calculados
resumen_estudiantes = {}

for nombre, notas in calificaciones.items():
    print(nombre,notas)
    resumen_estudiantes[nombre] = {        
        "nota_máxima": max(notas),
        "nota_mínima": min(notas),
        "promedio": sum(notas) / len(notas)
    }
    print(resumen_estudiantes[nombre])
    print("="*25)
```

```python
ranking = []
for nombre, stats in resumen_estudiantes.items():
    ranking.append((nombre, stats["promedio"]))

print(ranking)
```

```python
sorted(ranking, key=lambda x: x[1], reverse=True)
print(ranking)
```

```python
ranking[2:] = sorted(ranking[2:], key=lambda x: x[0])
print(ranking)
```

```python
# Ejercicio 03

usuarios = { 
    " ANA_01 ": {"nombre": " ana torres ", "correo": " ANA@MAIL.COM ", "edad": "24"}, 
    "LUIS_02": {"nombre": "luis perez", "correo": "Luis@Mail.com", "edad": "treinta"}, 
    " carla_03": {"nombre": "CARLA RUIZ", "correo": "carla@mail.com", "edad": "27"} 
}

usuarios_limpios = {}

for usuario, datos in usuarios.items():
    print(usuario, datos)

    # 1. Limpiar espacios en la usuario
    llave_limpia = usuario.strip()
   
    try:
        edad_int = int(datos["edad"])
    except ValueError:
        edad_int = None   

    usuarios_limpios[llave_limpia] = {
        "nombre": datos["nombre"].strip().title(),
        "correo": datos["correo"].strip().lower(),
        "edad": edad_int
    }

    print(usuario, usuarios_limpios[llave_limpia])
    print("=" * 80)
```

```python
# Ejercicio 04

texto = "Python es poderoso; Python es flexible, y aprender Python puede ser muy divertido."
palabras_vacias = {"es", "y", "ser", "muy", "puede"}

contador ={}

for palabra in texto.split(" "):
    palabra_limpia = palabra.lower().replace(";","").replace(",","").replace(".","")

    if palabra_limpia in palabras_vacias:
        continue

contador.setdefault(palabra_limpia,0)
contador[palabra_limpia] +=1

print(contador)
```

```python
# Ejercicio 05

empleadoleados = [
    {"nombre": "Ana", "departamento": "Desarrollo", "nivel": "Senior", "salario": 5200},
    {"nombre": "Luis", "departamento": "Diseño", "nivel": "Junior", "salario": 2800},
    {"nombre": "Carla", "departamento": "Desarrollo", "nivel": "Junior", "salario": 3400},
    {"nombre": "Pedro", "departamento": "Desarrollo", "nivel": "Senior", "salario": 5600},
    {"nombre": "Lucía", "departamento": "Diseño", "nivel": "Senior", "salario": 4700}
]

# Diccionario para la agrupación
agrupacion = {}

for empleado in empleadoleados:
    departamento = empleado["departamento"]
    nivel = empleado["nivel"]
    
    # Inicializar llaves si no existen
    if departamento not in agrupacion:
        agrupacion[departamento] = {}
    if nivel not in agrupacion[departamento]:
        agrupacion[departamento][nivel] = {"nombres": [], "cantidad": 0, "salario_total": 0}
    
    # Acumular datos
    agrupacion[departamento][nivel]["nombres"].append(empleado["nombre"])
    agrupacion[departamento][nivel]["cantidad"] += 1
    agrupacion[departamento][nivel]["salario_total"] += empleado["salario"]

# Calcular promedio y limpiar el total
for departamento in agrupacion:
    for nivel in agrupacion[departamento]:
        datos = agrupacion[departamento][nivel]
        datos["salario_promedio"] = datos["salario_total"] / datos["cantidad"]
        del datos["salario_total"]

print(agrupacion)
```

```python
# Ejercicio 06

tienda_a = {"laptop": 3500, "mouse": 80, "teclado": 160, "monitor": 920}
tienda_b = {"laptop": 3400, "mouse": 90, "teclado": 150, "webcam": 220}

# Obtener todos los productos únicos de ambas tiendas
productos = sorted(set(tienda_a.keys()) | set(tienda_b.keys()))

# imprimiendo cabecera
print(f"{'Producto':<12} | {'Tienda A':<8} | {'Tienda B':<8} | {'Más Barata':<10} | {'Diferencia'}")
print("-" * 65)

for p in productos:
    precio_a = tienda_a.get(p)
    precio_b = tienda_b.get(p)
    
    # comparación
    if precio_a and precio_b:
        mas_barata = "Tienda A" if precio_a < precio_b else "Tienda B"
        diferencia = abs(precio_a - precio_b)
        print(f"{p:<12} | {precio_a:<8} | {precio_b:<8} | {mas_barata:<10} | {diferencia}")
    
    # Manejo de productos exclusivos
    elif precio_a:
        print(f"{p:<12} | {precio_a:<8} | {'N/A':<8} | {'Tienda A':<10} | N/A")
    else:
        print(f"{p:<12} | {'N/A':<8} | {precio_b:<8} | {'Tienda B':<10} | N/A")
```

```python
# Ejercicio 07

# Lista de candidatos y votos
candidatos = ["Ana", "Luis", "Carla"]
votos = ["Ana", "Luis", "ana", "Carla", "Pedro", "Luis", "ANA", "", "Carla"]

mapa = {
    "ana": "Ana",
    "luis": "Luis",
    "carla": "Carla"
}


conteo = {}

for voto in votos:
    voto_limpio = voto.lower()
    candidato = mapa.get(voto_limpio, "invalido")

    if candidato == "invalido":
        conteo.setdefault("Votos invalidos", 0)
        conteo["Votos invalidos"] +=1
        continue

    conteo.setdefault(candidato, 0)
    conteo[candidato] +=1


print(conteo)

total_votos = len(votos)
for candidato in candidatos:
    conteo[f"{candidato}_pct"] = round(conteo[candidato] / total_votos * 100, 2)

```
