```python
# Ejercicio 01

numeros = (15, 8, 42, 3, 27, 19, 5)

mayor = max(numeros)
menor = min(numeros)
diferencia = mayor - menor

print(f"Mayor: {mayor}, Menor: {menor}, Diferencia: {diferencia}")
```

```python
# Ejercicio 02

datos = (5, 8, 5, 1, 9, 5, 2, 8, 5, 3)

numero = int(input("Introduce el número a buscar: "))
cantidad = datos.count(numero)

print(f"El número {numero} aparece {cantidad} veces.")
```

```python
# Ejercicio 03

nombres = ('Ana', 'Luis', 'Pedro', 'María', 'Luis', 'Carlos')
nombre_buscado = input("Introduce el nombre: ")

if nombre_buscado in nombres:
    posicion = nombres.index(nombre_buscado)
    print(f"El nombre '{nombre_buscado}' aparece por primera vez en la posición {posicion}.")
else:
    print("El nombre no existe en la tupla.")
```

```python
# Ejercicio 04

notas = (18, 15, 20, 11, 14, 17, 19)

maxima = max(notas)
minima = min(notas)
promedio = round(sum(notas) / len(notas),2)

# Contar aprobados (nota >= 11)
aprobados = 0
for nota in notas:
    if nota >= 11:
        aprobados += 1

print(f"Promedio: {promedio}")
print(f"Máxima: {maxima}")
print(f"Mínima: {minima}")
print(f"Aprobados: {aprobados}")
```

```python
ventas = [('Lunes', 120), ('Martes', 150), ('Miércoles', 95), ('Jueves', 180), ('Viernes', 210)]

total = 0
mayor_venta = -1
dia_mejor = ""

for dia, monto in ventas:
    total += monto
    if monto > mayor_venta:
        mayor_venta = monto
        dia_mejor = dia

promedio = round(total / len(ventas),2)

print(f"Día con mayor venta: {dia_mejor} ({mayor_venta})")
print(f"Total vendido: {total}")
print(f"Promedio de venta: {promedio}")
```
