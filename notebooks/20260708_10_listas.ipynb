```python
# Definimos una lista de nombres
nombres = ["Ana", "Luis", "Carlos", "María"]

# Usamos un bucle for para imprimir cada elemento
for nombre in nombres:
    print(f"Hola, {nombre}")
```

```python
# Lista de precios
precios = [15.50, 20.00, 5.75, 12.00]

total = 0

print("Lista de precios:")
for precio in precios:
    print(f"- S/ {precio:.2f}")
    total += precio  # Acumulamos el precio en la variable total

print(f"\nEl total de la suma es: S/ {total:.2f}")
```
