```python
# Ejercicio 02

class Rectangulo:
    def __init__(self, base: float, altura: float) -> None:
        self.base = base
        self.altura = altura

    def area(self):
        return self.base * self.altura
    
    def perimetro(self):
        return 2 * (self.base + self.altura)
    
    def __str__(self) -> str:
        return(f"Rectangulo de base: {self.base} y altura {self.altura}")
    

mi_rectangulo = Rectangulo(3,5)
print(mi_rectangulo.area())
print(mi_rectangulo.perimetro())

print(mi_rectangulo)
```

```python
# Ejercicio 05

class Libro:
    def __init__(self, titulo: str,autor: str,precio: float):
        self.titulo = titulo
        self.autor = autor
        self.__precio = precio

    def get_precio(self) -> float:
        return self.__precio
    
    def set_precio(self, nuevo_precio: float) -> float:
        if nuevo_precio > 0:
            self.__precio = nuevo_precio
        else:
            print(f"Error el nuevo precio es 0 o negativo: {nuevo_precio}")

libro = Libro(titulo = "Mi planta de naranja lima", autor ="Jose Mauro", precio = 20)
print(libro.get_precio())

libro.set_precio(50)
print(libro.get_precio()) 
```

```python
# Ejercicio 08

class Reproductor:
    def __init__(self):
        self.playlist = []
        self.posicion = 0

    def agregar_cancion(self, cancion: str) -> None:
        self.playlist.append(cancion)

    def cancion_actual(self) -> str:
        if not self.playlist:
            return "No hay canciones en la playlist"
        return self.playlist[self.posicion]
    
    def siguiente(self) ->str:
        self.posicion += 1
        if self.posicion < len(self.playlist):
            return self.cancion_actual()
        else:
            # cancion_actual = "No se puede reproducir la cancion siguiente"
            self.posicion -= 1
            return self.cancion_actual()

    def anterior(self) ->str:
        self.posicion -= 1
        if self.posicion >= 0:
            return self.cancion_actual()
        else:
            self.posicion += 1
            return self.cancion_actual()

    

mi_reproductor = Reproductor()
# mi_reproductor.cancion_actual()
mi_reproductor.agregar_cancion("I your self")
mi_reproductor.agregar_cancion("7 thing")
mi_reproductor.agregar_cancion("I need you")
mi_reproductor.siguiente()
```

```python
# Ejercicio 09

class Motor:
    def __init__(self, tipo: str, caballos: float) -> None:
        self.tipo = tipo
        self.caballos = caballos
        self.encendido = False

    def arrancar(self):
        if self.encendido:
            return "El motor ya está encendido"
        self.encendido = True
        return f"El motor {self.tipo} de {self.caballos} HP se encendió"
    
class Vehiculo:
    def __init__(self, marca: str, motor: Motor) -> None:
        self.marca = marca
        self.motor = motor

    def arrancar(self) -> str:
        return self.motor.arrancar()
    

motor = Motor("Gasolina", 150)
vehiculo = Vehiculo("Ford", motor)

print(vehiculo.arrancar())
print(vehiculo.arrancar())
        
```

```python
# Ejercicio 10

class Libro:
    def __init__(self, titulo: str, autor: str) -> None:
        self.titulo = titulo
        self.autor = autor
        self.disponible = True

class Usuario:
    def __init__(self, nombre: str) -> None:
        self.nombre = nombre
        self.libros_prestados = []

class Biblioteca:
    def __init__(self):
        self.libros = []

    def agregar_libro(self, libro: Libro) -> None:
        self.libros.append(libro)
    
    def buscar_libro(self, titulo: str) -> Libro | None:
        for libro in self.libros:
            if libro.titulo.lower() == titulo.lower():
                return libro
            
        return None
    
    def prestar_libro(self, titulo: str, usuario: Usuario) -> Libro | str:
        libro = self.buscar_libro(titulo)

        if libro is None:
            return "El libro no se encuentra en la biblioteca"
        
        if not libro.disponible:
            return "El libro ya fue prestado"
        
        libro.disponible = False
        usuario.libros_prestados.append(libro)

        return f"Libro prestado: {libro.titulo} al usuario: {usuario.nombre}"
    
    def devolver(self, titulo: str, usuario: Usuario):
        libro = self.buscar_libro(titulo)

        if libro is None:
            return "No existe el libro"
        
        if libro not in usuario.libros_prestados:
            return f"{usuario.nombre} no tiene prestado el libro: {titulo}"
        
        usuario.libros_prestados.remove(libro)
        libro.disponible = True

        return f"Libro: {titulo} fue devuelto por el usuario: {usuario.nombre}"
    
    def catalogo(self) -> None:
        for libro in self.libros:
            print(f"Titulo: {libro.titulo} | Autor: {libro.autor} | Estado: {libro.disponible}")


biblioteca = Biblioteca()
libro_1 = Libro("Mi planta de naranja lima","Jose Mauro")
libro_2 = Libro("Quien mató a palomino","Mario Vargas Llosa")
biblioteca.agregar_libro(libro_1)
biblioteca.agregar_libro(libro_2)

biblioteca.catalogo()

usuario_1 = Usuario("Pedro")
usuario_2 = Usuario("Sandra")

print("=" * 20)

print(biblioteca.prestar_libro("Mi planta de naranja lima", usuario_1))

print("=" * 20)

biblioteca.catalogo()

print("=" * 20)
for libro in usuario_1.libros_prestados:
    print(libro.titulo)

biblioteca.devolver("Mi planta de naranja lima", usuario_2)

print("=" * 20)
print(biblioteca.prestar_libro("La metamorfosis", usuario_1))
```
