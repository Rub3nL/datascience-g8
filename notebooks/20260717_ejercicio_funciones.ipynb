```python
# Ejercicio 01

def analizar_ventas(ventas: list[int | float]) -> dict[str, int | float]:

    if not ventas or any(venta < 0 for venta in ventas):
        return "Error: La lista está vacía o contiene valores negativos."
    
    total_vendido = sum(ventas)
    promedio = total_vendido / len(ventas)
    maxima = max(ventas)
    minima = min(ventas)
    
    # Conteo
    contar = 0
    for venta in ventas:
        if venta > promedio:
            contar = contar + 1

    
    vent_sup_promedio = sum(1 for v in ventas if v > promedio)
    vent_menor_100 = sum(1 for v in ventas if v < 100)
    
    # Construcción del diccionario de resultados
    resumen = {
        "total": total_vendido,
        "promedio": promedio,
        "maxima": maxima,
        "minima": minima,
        "ventas_superior_promedio": vent_sup_promedio,
        "ventas_menor_100": vent_menor_100
    }
       
    return resumen


ventas = [150.50, 200, 80.90, 300, 175.50, 90]
print(analizar_ventas(ventas))

ventas = [150.50, 200,-100]
print(analizar_ventas(ventas))
```

```python
# Ejercicio 02

def validar_password(password: str) -> tuple[bool,list[str]]:
    errores = []
    
    # Longitud mínima
    if len(password) < 8:
        errores.append("El password tiene menos de 8 caracteres.")
    
    contiene_minuscula = False
    contiene_mayuscula = False
    contiene_numero = False
       
    for caracter in password:
        if caracter .islower():
            contiene_minuscula = True
            break

    for caracter in password:
        if caracter .isupper():
            contiene_mayuscula = True
            break

    for caracter in password:
        if caracter .isnumeric():
            contiene_numero = True
            break       

    # Mayúscula, minúscula y número
    if not contiene_mayuscula:
        errores.append("Debe contener al menos una letra mayúscula.")
    if not contiene_minuscula:
        errores.append("Debe contener al menos una letra minúscula.")
    if not contiene_numero:
        errores.append("Debe contener al menos un número.")
        
    # Símbolos requeridos
    simbolos = "@#$%"
    if not any(c in simbolos for c in password):
        errores.append("Debe incluir al menos uno de estos símbolos: @, #, $ o %.")
        
    # Sin espacios
    if ' ' in password:
        errores.append("No debe contener espacios.")
        
    es_valida = len(errores) == 0
    return es_valida, errores

# Datos de prueba
es_valida, errores = validar_password('Python123')
print(f"¿Es válida?: {es_valida}")
print(f"Errores: {errores}")
```

```python
# Ejercicio 03

def registrar_estudiante(estudiantes: dict[str | int | float | str], nombre: str, notas: float) -> dict[str, int | float | str]:
    # Validar nombre vacío
    if not nombre or nombre.strip() == "":
        return "Error: El nombre no puede estar vacío."
    
    # Validar nombre duplicado
    if nombre in estudiantes:
        return f"Error: El estudiante {nombre} ya existe."
    
    # Validar cantidad de notas
    if len(notas) < 3:
        return "Error: El estudiante debe tener al menos tres notas."
    
    # Validar notas numéricas y rango (0-20)
    for nota in notas:
        # Primero validamos si es numérico
        if not isinstance(nota, (int, float)):
            return f"Error: La nota '{nota}' no es un valor numérico."
        
        # Luego validamos si está en el rango permitido
        if not (0 <= nota <= 20):
            return f"Error: La nota '{nota}' está fuera del rango permitido (0-20)."
    
    # Calcular promedio y estado
    promedio = sum(notas) / len(notas)
    estado = "Aprobado" if promedio >= 11 else "Desaprobado"
    
    # Guardar en el diccionario
    estudiantes[nombre] = {
        "notas": notas,
        "promedio": round(promedio, 2),
        "estado": estado
    }
    
    return f"Registro exitoso: {nombre} ha sido registrado como {estado}."
```

```python
# Ejercicio 04

def agrupar_por_longitud(palabras: str) -> dict[str, int | float | str]:
    resultado = {}
    vistas = set()  # Para rastrear duplicados (normalizados a minúsculas)

    for p in palabras:
        # Ignorar elementos que no sean cadenas
        if not isinstance(p, str):
            continue
        
        # Normalizar para evitar duplicados
        p_normalizada = p.lower()
        if p_normalizada in vistas:
            continue
        
        # Agregar a la estructura
        longitud = len(p)
        if longitud not in resultado:
            resultado[longitud] = []
        
        resultado[longitud].append(p)
        vistas.add(p_normalizada)
    
    # Ordenar alfabéticamente cada lista
    for longitud in resultado:
        resultado[longitud].sort()
        
    return resultado
```

```python
# Datos de prueba
palabras = ['python', 'Java', 'código', 'web', 'java', 'datos', 'API', 'lista']
print(agrupar_por_longitud(palabras))
```

```python
# Ejercicio 05

def actualizar_inventario(inventario: list, producto: str, cantidad: int, operacion: str):
    # Validar que la operación sea permitida
    if operacion not in ['entrada', 'salida']:
        return {"mensaje": "Error: La operación debe ser 'entrada' o 'salida'.", "stock_actual": None}
    
    # Validar que la cantidad sea un entero positivo
    if not isinstance(cantidad, int) or cantidad <= 0:
        return {"mensaje": "Error: La cantidad debe ser un entero positivo.", "stock_actual": None}
    
    # Procesar entrada / salida
    if operacion == 'entrada':
        inventario[producto] = inventario.get(producto, 0) + cantidad
        return {"mensaje": f"Entrada registrada: {producto}.", "stock_actual": inventario[producto]}
 
    elif operacion == 'salida':
        if producto not in inventario:
            return {"mensaje": f"Error: El producto '{producto}' no existe en el inventario.", "stock_actual": None}
        
        if inventario[producto] < cantidad:
            return {"mensaje": f"Error: Stock insuficiente para '{producto}'.", "stock_actual": inventario[producto]}
        
        inventario[producto] -= cantidad
        return {"mensaje": f"Salida registrada: {producto}.", "stock_actual": inventario[producto]}
```

```python
inventario = { 'Laptop': 5, 'Mouse': 12, 'Teclado': 8 }
print(actualizar_inventario(inventario, 'Mouse', 4, 'salida'))
print(f"Inventario final: {inventario}")
```

```python
# Ejercicio 06

def calcular_pedido(productos: list, cantidades: list, descuento: float, delivery: float):
    # Validación del descuento
    if not (0 <= descuento <= 100):
        return "Error: El descuento debe estar entre 0 y 100."
    
    subtotal = 0
    no_encontrados = []
    
    # Procesar el pedido
    for producto, cantidad in cantidades.items():
        # Validar que la cantidad sea un entero positivo
        if not isinstance(cantidad, int) or cantidad <= 0:
            continue
            
        # Verificar si existe en el catálogo
        if producto in productos:
            subtotal += productos[producto] * cantidad
        else:
            no_encontrados.append(producto)
            
    # Calcular montos finales
    monto_descuento = subtotal * (descuento / 100)
    total_final = subtotal - monto_descuento + delivery
    
    # Retornar resultados
    return {
        "subtotal": subtotal,
        "descuento": monto_descuento,
        "delivery": delivery,
        "total_final": total_final,
        "no_encontrados": no_encontrados
    }
```

```python

productos = {'Hamburguesa': 15, 'Pizza': 25, 'Gaseosa': 5}
cantidades = {'Hamburguesa': 2, 'Gaseosa': 3, 'Helado': 1}
descuento = 10
delivery = 5

resultado = calcular_pedido(productos, cantidades, descuento, delivery)
print(resultado)
```

```python
# Ejercicio 07

def obtener_clasificacion(promedio: float) -> str:
    if promedio >= 90: return "Excelente"
    if promedio >= 80: return "Bueno"
    if promedio >= 70: return "Regular"
    return "Deficiente"

def evaluar_empleados(empleados) -> list:
    resultados = []
    
    for nombre, puntajes in empleados.items():
        if not puntajes:
            continue
        
        promedio = sum(puntajes) / len(puntajes)
        
        if promedio >= 70:
            clasificacion = obtener_clasificacion(promedio)
            resultados.append((nombre, round(promedio, 2), clasificacion))
            
    # Ordenar de mayor a menor
    resultados.sort(key=lambda x: (-x[1], x[0]))
    
    return resultados
```

```python

empleados = { 
    'Ana': [80, 90, 75], 
    'Carlos': [60, 70, 65], 
    'Lucía': [95, 85, 90], 
    'Pedro': [70, 70, 70] 
}
print(evaluar_empleados(empleados))
```

```python
# Ejercicio 08

def comprimir_texto(texto: str) -> str:
    """Comprime texto reemplazando caracteres consecutivos por el carácter y su cantidad."""
    if not texto:
        return ""
    
    resultado: str = ""
    contador: int = 1
    
    for i in range(1, len(texto)):
        if texto[i] == texto[i - 1]:
            contador += 1
        else:
            resultado += f"{texto[i - 1]}{contador}"
            contador = 1
    
    # Agregar el último grupo procesado
    resultado += f"{texto[-1]}{contador}"
    return resultado

def descomprimir_texto(texto: str) -> str:
    """Reconstruye el texto original, admitiendo cantidades de más de un dígito."""
    if not texto:
        return ""
    
    resultado: str = ""
    i: int = 0
    
    while i < len(texto):
        # 1. Validar que el primer elemento sea una letra
        caracter: str = texto[i]
        if not caracter.isalpha():
            return "Error: Formato comprimido inválido."
        i += 1
        
        # 2. Capturar el número (puede tener uno o más dígitos)
        num_str: str = ""
        while i < len(texto) and texto[i].isdigit():
            num_str += texto[i]
            i += 1
            
        # Si no hay número tras la letra, el formato es incorrecto
        if not num_str:
            return "Error: Formato comprimido inválido."
            
        # 3. Expandir el carácter según la cantidad encontrada
        resultado += caracter * int(num_str)
        
    return resultado
```

```python
texto_original = 'aaabbccccdaa'
comprimido = comprimir_texto(texto_original)
descomprimido = descomprimir_texto(comprimido)

print(f"Original: {texto_original}")
print(f"Comprimido: {comprimido}")
print(f"Descomprimido: {descomprimido}")
```

```python
# Ejercicio 09

def transferir(cuentas, origen, destino, monto):
    # Verificar que las cuentas de origen y destino existan
    if origen not in cuentas or destino not in cuentas:
        return {"estado": "Fallido", "error": "Una o ambas cuentas no existen."}

    # Las cuentas deben ser diferentes
    if origen == destino:
        return {"estado": "Fallido", "error": "La cuenta de origen y destino no pueden ser iguales."}

    # El monto debe ser numérico y mayor que cero
    if not isinstance(monto, (int, float)) or monto <= 0:
        return {"estado": "Fallido", "error": "El monto debe ser un número positivo."}

    # La cuenta de origen debe tener saldo suficiente
    if cuentas[origen]['saldo'] < monto:
        return {"estado": "Fallido", "error": "Saldo insuficiente en cuenta de origen."}

    # Si todo es correcto, actualizar ambos saldos
    cuentas[origen]['saldo'] -= monto
    cuentas[destino]['saldo'] += monto

    # Retornar comprobante con los datos solicitados
    return {
        "estado": "Éxito",
        "titular_origen": cuentas[origen]['titular'],
        "titular_destino": cuentas[destino]['titular'],
        "monto_transferido": monto,
        "saldo_restante": cuentas[origen]['saldo']
    }
```

```python

cuentas = { 
    '001': {'titular': 'Ana', 'saldo': 1500}, 
    '002': {'titular': 'Luis', 'saldo': 800} 
}

print(transferir(cuentas, '001', '002', 300))
```

```python
def mostrar_sala(sala):    
    print("Estado actual de la sala:")
    print("   " + " ".join([str(i) for i in range(len(sala[0]))]))
    for i, fila in enumerate(sala):
        print(f"{i}  " + " ".join(fila))
    print()

def reservar_asientos(sala, asientos_solicitados):
    # Validar que todas las posiciones existan y estén libres
    for fila, col in asientos_solicitados:
        # Validar límites de la lista
        if fila < 0 or fila >= len(sala) or col < 0 or col >= len(sala[0]):
            return {"estado": "Error", "mensaje": f"El asiento ({fila}, {col}) no existe."}
        
        # Validar disponibilidad
        if sala[fila][col] == 'O':
            return {"estado": "Error", "mensaje": f"El asiento ({fila}, {col}) ya está ocupado."}

    # Si todo es correcto, realizar la reserva
    for fila, col in asientos_solicitados:
        sala[fila][col] = 'O'
    
    return {
        "estado": "Éxito",
        "asientos_reservados": asientos_solicitados,
        "sala": sala
    }
```

```python

sala = [['L', 'L', 'O', 'L'], ['O', 'L', 'L', 'L'], ['L', 'O', 'L', 'L']]
asientos = [(1, 2), (2, 3), (0, 0)]

mostrar_sala(sala)
resultado = reservar_asientos(sala, asientos)

if resultado["estado"] == "Éxito":
    print("¡Reserva realizada con éxito!")
    mostrar_sala(resultado["sala"])
else:
    print(f"Error en la reserva: {resultado['mensaje']}")
```
