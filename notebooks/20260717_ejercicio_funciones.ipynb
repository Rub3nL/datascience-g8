```python
# Ejercicio 01

def analizar_ventas(ventas: list[int | float]) -> dict[str, int | float]:

    if not ventas or any(venta < 0 for venta in ventas):
        return "Error: La lista está vacía o contiene valores negativos."
    
    total_vendido = sum(ventas)
    promedio = total_vendido / len(ventas)
    maxima = max(ventas)
    minima = min(ventas)
    
    # Conteo
    contar = 0
    for venta in ventas:
        if venta > promedio:
            contar = contar + 1

    
    vent_sup_promedio = sum(1 for v in ventas if v > promedio)
    vent_menor_100 = sum(1 for v in ventas if v < 100)
    
    # Construcción del diccionario de resultados
    resumen = {
        "total": total_vendido,
        "promedio": promedio,
        "maxima": maxima,
        "minima": minima,
        "ventas_superior_promedio": vent_sup_promedio,
        "ventas_menor_100": vent_menor_100
    }
       
    return resumen


ventas = [150.50, 200, 80.90, 300, 175.50, 90]
print(analizar_ventas(ventas))

ventas = [150.50, 200,-100]
print(analizar_ventas(ventas))
```

```python
# Ejercicio 02

def validar_password(password: str) -> tuple[bool,list[str]]:
    errores = []
    
    # 1. Longitud mínima
    if len(password) < 8:
        errores.append("El password tiene menos de 8 caracteres.")
    
    contiene_minuscula = False
    contiene_mayuscula = False
    contiene_numero = False
       
    for caracter in password:
        if caracter .islower():
            contiene_minuscula = True
            break

    for caracter in password:
        if caracter .isupper():
            contiene_mayuscula = True
            break

    for caracter in password:
        if caracter .isnumeric():
            contiene_numero = True
            break       

    # 2. Mayúscula, minúscula y número
    if not contiene_mayuscula:
        errores.append("Debe contener al menos una letra mayúscula.")
    if not contiene_minuscula:
        errores.append("Debe contener al menos una letra minúscula.")
    if not contiene_numero:
        errores.append("Debe contener al menos un número.")
        
    # 3. Símbolos requeridos
    simbolos = "@#$%"
    if not any(c in simbolos for c in password):
        errores.append("Debe incluir al menos uno de estos símbolos: @, #, $ o %.")
        
    # 4. Sin espacios
    if ' ' in password:
        errores.append("No debe contener espacios.")
        
    es_valida = len(errores) == 0
    return es_valida, errores

# Datos de prueba
es_valida, errores = validar_password('Python123')
print(f"¿Es válida?: {es_valida}")
print(f"Errores: {errores}")
```

```python
# Ejercicio 03

def registrar_estudiante(estudiantes: dict[str | int | float | str], nombre: str, notas: float) -> dict[str, int | float | str]:
    # 1. Validar nombre vacío
    if not nombre or nombre.strip() == "":
        return "Error: El nombre no puede estar vacío."
    
    # 2. Validar nombre duplicado
    if nombre in estudiantes:
        return f"Error: El estudiante {nombre} ya existe."
    
    # 3. Validar cantidad de notas
    if len(notas) < 3:
        return "Error: El estudiante debe tener al menos tres notas."
    
    # 4. Validar notas numéricas y rango (0-20)
    for nota in notas:
        # Primero validamos si es numérico
        if not isinstance(nota, (int, float)):
            return f"Error: La nota '{nota}' no es un valor numérico."
        
        # Luego validamos si está en el rango permitido
        if not (0 <= nota <= 20):
            return f"Error: La nota '{nota}' está fuera del rango permitido (0-20)."
    
    # 5. Calcular promedio y estado
    promedio = sum(notas) / len(notas)
    estado = "Aprobado" if promedio >= 11 else "Desaprobado"
    
    # 6. Guardar en el diccionario
    estudiantes[nombre] = {
        "notas": notas,
        "promedio": round(promedio, 2),
        "estado": estado
    }
    
    return f"Registro exitoso: {nombre} ha sido registrado como {estado}."
```

```python
estudiantes = {}
print(registrar_estudiante(estudiantes, 'Marco', [10, 5, 12]))
print(estudiantes)
```
