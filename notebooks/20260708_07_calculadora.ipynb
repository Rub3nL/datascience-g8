```python
cuenta = float(input("Ingresa el total de la cuenta: "))

porcentaje = float(input("Ingrese el % de la propina: "))

propina = cuenta * (porcentaje / 100)

print(f"El importe de la propina es : S/ { round(propina,2)}")
print(f"El importe de la total es : S/ {round(cuenta + propina,2)}")

```
